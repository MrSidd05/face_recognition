```python
import os
import cv2
import face_recognition
import numpy as np
import telepot
import time

# Set up the Telegram bot
bot = telepot.Bot('5937240704:AAFlu3GIOY0sf8onY12ECTpYXfd4Vl6B_Ro')
chat_id = '1019525030'

# Load the images and corresponding class names
path = '../ImagesAttendence'
images = []
classNames = []
mylist = os.listdir(path)

for cl in mylist:
    curImg = cv2.imread(f'{path}/{cl}')
    images.append(curImg)
    classNames.append(os.path.splitext(cl)[0])

# Encode the known faces
def findEncodings(images):
    encodeList = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        encodeList.append(encode)
    return encodeList

encodelistknown = findEncodings(images)

# Set up the webcam capture
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FOCUS, 0)
new_face_detected = False

while True:
#     time.sleep(2)
    success, img = cap.read()
    imgS = cv2.resize(img, (0, 0), None, 0.25, 0.25)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

    # Find the locations and encodings of faces in the current frame
    facecurFrame = face_recognition.face_locations(imgS)
    encodecurFrame = face_recognition.face_encodings(imgS)

    for encodeFace, faceLoc in zip(encodecurFrame, facecurFrame):
        matches = face_recognition.compare_faces(encodelistknown, encodeFace)
        faceDis = face_recognition.face_distance(encodelistknown, encodeFace)
        matchIndex = np.argmin(faceDis)

        if matches[matchIndex]:
            name = classNames[matchIndex].upper()
            y1, x2, y2, x1 = faceLoc
            y1, x2, y2, x1 = y1*4, x2*4, y2*4, x1*4
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(img, name, (x1+6, y2+6), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
            new_face_detected = False
        else:
            if not new_face_detected:
                name = "unknown"
                y1, x2, y2, x1 = faceLoc
                y1, x2, y2, x1 = y1*4, x2*4, y2*4, x1*4
                cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(img, name, (x1+6, y2+6), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
                current_time = time.strftime('%H:%M:%S', time.localtime())
                time.sleep(3)
                cv2.imwrite('new_face.jpg', img[y1:y2, x1:x2])
                print(f"New face detected at {current_time}")
                with open('new_face.jpg', 'rb') as f:
                    bot.sendPhoto(chat_id, f, caption=f"New face detected at {current_time}")
                os.remove('new_face.jpg')
                new_face_detected = True
            else:
                pass

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
